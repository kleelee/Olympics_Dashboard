```python
df = pd.read_csv("athlete_events.csv")
```

```python
#Create a table grouped by Team, Year, and Medal to get medal counts.
countrygroupdf = df.groupby(["Team","Year","Medal"]).size()
```

```python
countrydf = pd.DataFrame(countrygroupdf).reset_index()
```

```python
#Create a table grouped by Team, Year, and Medal to get medal counts.
namegroupdf = df.groupby(["Name","Sex","Team","Sport","Medal"]).size()
```

```python
namedf = pd.DataFrame(namegroupdf).reset_index()
```

```python
namedf.head()
```

```python
goldathletes = namedf[namedf["Medal"] == "Gold"].reset_index()

del goldathletes["index"]
goldathletesdf = goldathletes.rename(columns={0:"Medal_Count"})

goldathletesdf.sort_values(by=["Medal_Count"], ascending=False)
```
